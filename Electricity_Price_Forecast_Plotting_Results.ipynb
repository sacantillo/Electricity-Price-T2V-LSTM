```python
from scipy.stats import wilcoxon
diferencias1 = TRF.values[:,3]
diferencias2 = HW.Errors.values[31:]

# Realiza la prueba de Wilcoxon
statistic, p_value = wilcoxon(diferencias1,diferencias2)

# Imprime los resultados
print("Estadístico de la prueba:", statistic)
print("Valor p:", p_value)

# Compara el valor p con el nivel de significancia
nivel_significancia = 0.05
if p_value < nivel_significancia:
    print("Se rechaza la hipótesis nula: Hay una diferencia significativa.")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa.")
```
